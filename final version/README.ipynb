## 初始化流程
1. **配置数据库连接**

   打开项目文件：
   mybackend/mybackend/settings.py
   
修改 `DATABASES` 配置为你本地 MySQL 的信息(数据库名称以及本机密码)，示例如下：

    ```python
    DATABASES = {
        'default': {
            'ENGINE': 'django.db.backends.mysql',
            'NAME': 'job_platform',
            'USER': 'root',
            'PASSWORD': 'database_password',
            'HOST': 'localhost',
            'PORT': '3306',
            'OPTIONS': {
                'charset': 'gbk',
                'init_command': "SET sql_mode='STRICT_TRANS_TABLES'"
            },
        }
    }
    ```


2. **配置数据库连接**

    pip install mysqlclient

    pip install django-nested-admin
    
    pip install djangorestframework

3. **生成并应用数据迁移**(如果使用导入的测试数据可以无视这一栏)

    python manage.py makemigrations

    python manage.py migrate

## 测试数据导入方法

1.先下载压缩包中的data.sql文件

2.打开本机的Mysql WorkBench

3.进入主界面,点击导航栏Server中的Data Import

4.选择Import From self-contained file,输入你下载好的data.sql的地址

5.更改新建databse为job_platform.  注:如果改名,需要在mybackend/mybackend/settings.py修改数据库名称

5.点击右下角start import即可完成导入数据
